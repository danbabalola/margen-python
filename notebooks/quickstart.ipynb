Pull labeled deepfake attack-data (real vs AI-generated faces) by type. A **test** key pulls a free fixed sample; a **live** key debits one credit per image. Get a key at [www.margensoftware.com/keys](https://www.margensoftware.com/keys).

```python
from margen import Margen
from margen.ergonomics import iter_items, iter_lineages, download_selection

# Paste your key (test keys start with mgn_test_)
client = Margen(bearer_auth="mgn_test_...")

# Which benchmarks can this key query?
for b in client.list_benchmarks().data:
    print(b.id, "-", b.title)
```

```python
# Select the dark-cell fakes from synthetic-face-v1 (a list value ORs; omit a filter for all)
picks = list(iter_items(client, benchmark="synthetic-face-v1", kind="fake", skin_tone="dark"))
print(len(picks), "items")
```

```python
# Download the whole selection to a folder (test tier is free; live debits 1 credit/image)
saved = download_selection(client, picks, "margen_out")
print("saved", len(saved), "images to margen_out/")
```

```python
# Preview one
from IPython.display import Image, display
if saved:
    display(Image(filename=saved[0]))
```
